```python
# Remove commas and convert 'Density P/km^2' column to numeric values for sorting
cleaned_population_data['Density P/km^2'] = pd.to_numeric(cleaned_population_data['Density P/km^2'])

# Sort the DataFrame by density in descending order and select the top 10 countries
top_10_dense_countries = cleaned_population_data.nlargest(10, 'Density P/km^2')

# Sort the DataFrame by density in ascending order to get the 10 least crowded countries
least_crowded_countries = cleaned_population_data.nsmallest(10, 'Density P/km^2')

# Print the top 10 crowded countries
print("Top 10 Crowded Countries:")
print(top_10_dense_countries)

# Print the 10 least crowded countries
print("\n10 Least Crowded Countries:")
print(least_crowded_countries)
```

```python
# Remove commas from the 'Migrants_net' column and convert it to numeric values for sorting
cleaned_population_data['Migrants_net'] = pd.to_numeric(cleaned_population_data['Migrants_net'])

# Sort the DataFrame by 'Migrants_net' in descending order to get the top 10 countries with the highest migrants
highest_migrant_countries = cleaned_population_data.nlargest(10, 'Migrants_net')

# Sort the DataFrame by 'Migrants_net' in ascending order to get the 10 countries with the lowest migrants
lowest_migrant_countries = cleaned_population_data.nsmallest(10, 'Migrants_net')

# Print the top 10 countries with the highest migrants
print("Top 10 Countries with the Highest Migrants:")
print(highest_migrant_countries)

# Print the 10 countries with the lowest migrants
print("\n10 Countries with the Lowest Migrants:")
print(lowest_migrant_countries)
```

```python
# Convert 'Fertility_Rate' column to numeric values for sorting
cleaned_population_data['Fertility_Rate'] = pd.to_numeric(cleaned_population_data['Fertility_Rate'])

# Sort the DataFrame by 'Fertility_Rate' in descending order to get the top 10 countries with the highest fertility rates
highest_fertility_countries = cleaned_population_data.nlargest(10, 'Fertility_Rate')

# Sort the DataFrame by 'Fertility_Rate' in ascending order to get the 10 countries with the lowest fertility rates
lowest_fertility_countries = cleaned_population_data.nsmallest(10, 'Fertility_Rate')

# Print the top 10 countries with the highest fertility rates
print("Top 10 Countries with the Highest Fertility Rates:")
print(highest_fertility_countries)

# Print the 10 countries with the lowest fertility rates
print("\n10 Countries with the Lowest Fertility Rates:")
print(lowest_fertility_countries)
```

```python
import ipywidgets as widgets
from IPython.display import display, IFrame
import pandas as pd
import matplotlib.pyplot as plt

# Load the data frames
# Define the data frames as you mentioned in your initial request

data_frames = {
    'High Population Countries': (high_population_countries, high_population_map_path),
    'Lowest Population Countries': (lowest_population_countries, lowest_population_map_path),
    'Top 10 Dense Countries': (top_10_dense_countries, top_dense_countries_map_path),
    'Least Crowded Countries': (least_crowded_countries, least_crowded_countries_map_path),
    'Highest Migrant Countries': (highest_migrant_countries, highest_migrant_countries_map_path),
    'Lowest Migrant Countries': (lowest_migrant_countries, lowest_migrant_countries_map_path),
    'Highest Fertility Countries': (highest_fertility_countries, highest_fertility_countries_map_path),
    'Lowest Fertility Countries': (lowest_fertility_countries, lowest_fertility_countries_map_path),
    'Oldest Countries': (oldest_countries, oldest_countries_map_path),
    'Youngest Countries': (youngest_countries, youngest_countries_map_path)
}

def display_data_frame_and_map(data_frame_name):
    data_frame, map_path = data_frames[data_frame_name]
    
    # Display the data frame
    print(f"Data Frame: {data_frame_name}")
    display(data_frame)
    
    # Display the map using IFrame
    print(f"Map for {data_frame_name}:")
    display(IFrame(src=map_path, width='100%', height=500))

data_frame_dropdown = widgets.Dropdown(
    options=list(data_frames.keys()),
    description='Select Data Frame:',
    disabled=False,
)

widgets.interactive(display_data_frame_and_map, data_frame_name=data_frame_dropdown)
```
