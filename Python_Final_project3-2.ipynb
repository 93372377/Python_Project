```python
import numpy as np
import pandas as pd

def clean_data():
    global combined_data 

 
    combined_data.replace('N.A.', np.nan, inplace=True)

    
    cleaned_combined_data = combined_data.dropna()

    return cleaned_combined_data

cleaned_combined_data = clean_data()
```

```python
import pandas as pd

def get_highest_population_countries(cleaned_combined_data, country_iframes):
    cleaned_combined_data['Population 2023'] = pd.to_numeric(cleaned_combined_data['Population 2023'], errors='coerce').fillna(0)
    top_population_countries = cleaned_combined_data.nlargest(10, 'Population 2023').copy()
    
    display(show_population_map(top_population_countries, country_iframes))
    
    print("\nTop 10 Countries with Highest Population:")
    print(top_population_countries)
    
    return top_population_countries

def get_lowest_population_countries(cleaned_combined_data, country_iframes):
    cleaned_combined_data['Population 2023'] = pd.to_numeric(cleaned_combined_data['Population 2023'], errors='coerce').fillna(0)
    lowest_population_countries = cleaned_combined_data.nsmallest(10, 'Population 2023').copy()
   
    display(show_population_map(lowest_population_countries, country_iframes))
    
    print("\nTop 10 Countries with Lowest Population:")
    print(lowest_population_countries)
    
    return lowest_population_countries

high_population_countries = get_highest_population_countries(cleaned_combined_data.copy(), country_iframes)
low_population_countries = get_lowest_population_countries(cleaned_combined_data.copy(), country_iframes)
```

```python
import ipywidgets as widgets
from IPython.display import display

def create_dashboard():
    
    criteria_dropdown = widgets.Dropdown(
        options=['High Population', 'Low Population', 'Top Crowded', 'Least Crowded', 'High Migrant', 
                 'Low Migrant', 'High Fertility', 'Low Fertility', 'Oldest', 'Youngest', 'Richest', 'Most Poor'],
        description='Select Criteria:'
    )

    display(criteria_dropdown)

    output_container = widgets.Output()
    display(output_container)

    def display_data(criteria):
        output_container.clear_output()
        with output_container:
            if criteria == 'High Population':
                data = high_population_countries
            elif criteria == 'Low Population':
                data = low_population_countries
            elif criteria == 'Top Crowded':
                data = top_crowded_countries
            elif criteria == 'Least Crowded':
                data = least_crowded_countries
            elif criteria == 'High Migrant':
                data = top_migrant_countries
            elif criteria == 'Low Migrant':
                data = least_migrant_countries
            elif criteria == 'High Fertility':
                data = top_fertility_countries
            elif criteria == 'Low Fertility':
                data = lowest_fertility_countries
            elif criteria == 'Oldest':
                data = oldest_countries
            elif criteria == 'Youngest':
                data = youngest_countries
            elif criteria == 'Richest':
                data = richest_countries
            elif criteria == 'Most Poor':
                data = most_poor_countries

            display(data[['Country']])

            display(show_population_map(data, country_iframes))  

    def on_dropdown_change(change):
        if change['type'] == 'change' and change['name'] == 'value':
            display_data(change['new'])

    criteria_dropdown.observe(on_dropdown_change)

    display_data(criteria_dropdown.value)

create_dashboard()
```
